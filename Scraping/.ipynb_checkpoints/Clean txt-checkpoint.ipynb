Most of the MetMuseum Collection Catalogues and exhibition texts were available as PDF files at 

```python
import os
import re
import glob
```

```python
#Insert directory you're working on into the parentheses after glob and before *.txt
book_filenames = sorted(glob.glob(r"D:\ML\Training-the-Archive\trainingdata.txt"))
print("Found books:")
book_filenames
```

```python
for book_filename in book_filenames: 
        chapapelo = open(book_filename, "r", encoding='charmap')
        coso = chapapelo.read()
        #this pastes together all line breaks, the cause of most of the problems
        article_text = re.sub(r', ', '_', coso)
        article_text = re.sub(r'D:\\ML\\Training-the-Archive\\Met_Collection\\', '/content/drive/My Drive/Colab Notebooks/Training-the-Archive/Met_Collection/', article_text)
        article_text = re.sub(r'/content/drive/My Drive/Colab Notebooks/Training-the-Archive/Met_Collection/', '', article_text)
        article_text = re.sub(r" ", "_", article_text)
        article_text = re.sub(r"'", "_", article_text)
        #article_text = re.sub(r".jpg", ".jpg,", article_text)
#         article_text = re.sub(r'‘', ' ', article_text)
#         article_text = re.sub(r'frangais', 'francais', article_text)
#         article_text = re.sub(r'ó', 'o', article_text)
#         article_text = re.sub(r'ú', 'u', article_text)
#         article_text = re.sub(r'veldzques', 'velazques', article_text)
#         #This line gets rid of any character that is not alphanumerical or , and .
#         article_text = re.sub(r'[^a-zA-Z0-9,.]', ' ', article_text)
#         #Gets rid of all numbers
#         article_text = re.sub(r'\d+', ' ', article_text)
#         #normalizes all espaces to only one
#         article_textclean = re.sub(r'\s+', ' ', article_text)
        
        #Creates a new directory within folder where to dump all newly created files
        file_name = os.path.basename(book_filename)[:-4] + "_clean.txt"
        script_dir = os.path.dirname(os.path.abspath(book_filename))
        dest_dir = os.path.join(script_dir, 'CleanBooks2')
        try:
            os.makedirs(dest_dir)
        except OSError:
            pass # already exists
        path = os.path.join(dest_dir, file_name)
        
        with open(path, 'w') as stream:
            stream.write(article_text)#clean)
            stream.close()
        print("processed:" + file_name)
```
