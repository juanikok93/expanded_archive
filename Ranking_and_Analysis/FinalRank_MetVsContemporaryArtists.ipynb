## Methods for assembling the ranking

The rankings assembled have different lenghts. The text ranking, has values between 1-52. The image ranking has values between 1-8743. For all values to have a comparable measurement, i took the following steps:

- Averaging frequency list and tf-idf for text distances.
- Normalize using min-max scaler (lowest value 0 and highest value 1) the similarities and distances of both text and image
- Parse the text values to each image row merging by artist name. In this way, each image is associated to the same artist text.
- Make a ranking averaging the scaled values for: 1) Text similarity to all keywords; 2) Text similarity to closest 30 keywords; 3) Image similarity to all Metmuseum Collection, 4) Image similarity to top 10 images.

The assumption is that the image with the lowest number (distance) is the one that finds itself closer to the MetMuseum collection and its exhibition catalogues, and thus "curated" by the algorithm as the most suitable for an exhibition.

*NORMALIZATION*

I apply normalization to the similarity values here, so they can be paired up with the distance values of the image dataframe.

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalingtxt = MinMaxScaler()
```

```python
scaled_txt_scores = scalingtxt.fit_transform(text_df[['Averaged simmil core', 'Averaged simmil top']])
```

```python
scaled_core_score = []
scaled_top_score = []

for x in scaled_txt_scores:
    scaled_core_score.append(x[0])
    scaled_top_score.append(x[1])
```

Since the text are *similarity* values and images are *distance* values, *I flip the polarity of the min-max*, like this the *higher number (most similar) would be 0* and the *lowest number (least similar) would be 1*. In the end, the results will be given in distance values*

```python
def flipping(X, lista):
    y = (max(lista) - X) / (max(lista) - min(lista))
    return y
```

```python
flipped_core = []
flipped_top = []

for x in scaled_core_score:
    y = flipping(x, scaled_core_score)
    flipped_core.append(y)

for x in scaled_top_score:
    y = flipping(x, scaled_top_score)
    flipped_top.append(y)
```

```python
text_df['Scaled Text Core Score'] = flipped_core
text_df['Scaled Text Top Score'] = flipped_top
```

### Conclusions

During this process of word vectorization, image embedding and distance calculation, we have operated within the realm of computability. The relationships that create meaning for us, humans, are not necessarily related only to simmilarity, or distance. Sometimes, there might be images that combined convey meaning, yet their relation is hard to write down in code, or to represent through vectors. 

This effort to curate a show based on text and image, and pursuing the parameter of "similarity" is but one way of organizing information, perhaps one of the least imaginative possible ways in curatorial terms. Yet, this process enables the processing of large scale datasets, in an automated process that goes over thousands of images and millions of words in a few hours. The processing scale is what emerges as a valuable tool for curators, yet is only one of the primary steps in order to assemble a meaningful show.

I went through hundreds of the results in the ranking, checking the image ranked, the words associated with it and the images it sits close to. In general terms, the program works given it prompts up images that formally resemble the MetMusem collection, with ocassional off-matches corresponding to the text simmilarity. Nevertheless, the literal final ranking could not be used in order to curate a show. 

To curate a real show, i took the names of artist atop of the list, and took a work which had interesting connections with the collection. After digging up more information, i was able to define a selection that was also based in affective links that emerged from the raw data.

This process will not replace human-based art curating, given that the purpose of most art currently produced is to inspire and stimulate humans, not machines. Yet, this process was able to open up the scope of data analyzed, and facilitated encounters that would have been impossible without automation.
